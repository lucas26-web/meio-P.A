```python
# operadores aritméticos

# fazer um programa que o usuário digita 02 números
# saída mostrar operações aritméticas

# entrada

a = int(input("digite um número ;)"))
b = int(input("digite um número ;)"))

# processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b

# saída

print("")
print("----------------------OPERADORES ARITMÉTICOS----------------------")

print("---------------------------SOMA-----------------------------------")
print("                                                                  ")
print(f"O resultado da SOMA de {a} mais {b} é igual a -> {soma}")
print("")
print("-----------------------------SUBTRAÇÃO-------------------------------")
print("")
print(f"O resultado da SUBTRAÇÃO  de {a} MENOS {b} é igual a -> {subtracao}")
print("")
print("-------------------------MULTIPLICAÇÃO-----------------------------")
print("")
print(f"O resultado da MULTIPLICAÇÃO de {a} vezes {b} é igual a -> {multiplicacao} ")
print("")
print("----------------------------DIVISÃO--------------------------------")
print("")
print(f"O resultado da DIVISÃO de {a} dividido {b} é igual a -> {divisao}")
print("")
print("-----------------------------DIV-INT-----------------------")
print("")
print(f"O resultado da DIVISÃO INDIRETA de {a} dividido {b} é igual a -> {div_int}")
print("")
print("-----------------------------RESTO-----------------------")
print("")
print(f"O resultado do RESTO de {a} dividido {b} é igua a -> {resto}")
print("")
print("-----------------------------POTÊNCIA-----------------------")
print("")
print(f"O resultado da POTÊNCIA de {a} elevado {b} é igual a -> {potencia}")
```
