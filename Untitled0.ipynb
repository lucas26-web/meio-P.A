```python
#questão 3

# escrever um programa que o usuário digita o nome e as suas 04 notas referente aos bimestres, calcula a média e mostra
#aprovado, rescuperação ou reprovado.

#nota maior ou igual a sete aprovado;
#nota maior ou igual 5 e menor 7 recuperação;
#nota menor 5 reprovado;

# entrada

nome = str (input("Digite seu nome: "))
nota1 = float (input("Digite sua nota do 1°  bimestre:  "))
nota2 = float (input("Digite sua nota do 2° bimestre:  "))
nota3 = float (input("Digite sua nota do 3°  bimestre:  "))
nota4 = float (input("Digite sua nota do 4°  bimestre:  "))


media = (nota1 + nota2 + nota3 + nota4) / 4


if media >= 7:
  print(f"{nome}, você foi aprovado com média {media}, parabéns!!")

elif media >= 5:
  print(f"{nome}, você está de recuperação, dá pra melhorar ;)")

else:
  print(f"{nome}, você foi reprovado.")
```

```python
#questão 4

#cria um programa que analisa idade e classifique entre criança, adolescente, adulto e idoso

# entrada

nome = str(input("Digite seu nome:  "))
idade = float(input("Digite sua idade:  "))

if idade <= 12:
  print(f"{nome} você é uma criança")

elif idade >= 13 and idade <= 17:
  print(f"{nome} você é um adolescente")

elif idade >= 18 and idade <= 60:
    print(f"{nome} você é um adulto")

elif idade >= 61 and idade <= 100:
  print(f"{nome} você é um idoso")
```

```python
#questão 7

# operadores aritméticos

# fazer um programa que o usuário digita 02 números
# saída mostrar operações aritméticas

# entrada

a = int(input("digite um número ;)"))
b = int(input("digite um número ;)"))

# processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b

# saída

print("")
print("----------------------OPERADORES ARITMÉTICOS----------------------")

print("---------------------------SOMA-----------------------------------")
print("                                                                  ")
print(f"O resultado da SOMA de {a} mais {b} é igual a -> {soma}")
print("")
print("-----------------------------SUBTRAÇÃO-------------------------------")
print("")
print(f"O resultado da SUBTRAÇÃO  de {a} MENOS {b} é igual a -> {subtracao}")
print("")
print("-------------------------MULTIPLICAÇÃO-----------------------------")
print("")
print(f"O resultado da MULTIPLICAÇÃO de {a} vezes {b} é igual a -> {multiplicacao} ")
print("")
print("----------------------------DIVISÃO--------------------------------")
print("")
print(f"O resultado da DIVISÃO de {a} dividido {b} é igual a -> {divisao}")
```

```python
# questão 8

# Pede um número de 1 a 7 e mostra o dia correspondente

# entrada
numero_dia = int(input("Digite um número de 1 a 7: "))

# processamento e saída
if numero_dia == 1:
    print("1 = Domingo")

elif numero_dia == 2:
    print("2 = Segunda-feira")

elif numero_dia == 3:
    print("3 = Terça-feira")

elif numero_dia == 4:
    print("4 = Quarta-feira")

elif numero_dia == 5:
    print("5 = Quinta-feira")

elif numero_dia == 6:
    print("6 = Sexta-feira")

elif numero_dia == 7:
    print("7 = Sábado")

else:
    print("Número inválido. Por favor, digite um número entre 1 e 7.")
```
