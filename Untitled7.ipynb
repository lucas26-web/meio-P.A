```python
# escrever um programa que recebe 04 notas do aluno, calcula a média e imprimi aprovado, recuperação ou reprovado

# regras:

# utilizar while
# nota maior ou igual a 7 aprovado
# nota maior ou igual a 5 recuperação
# nota menor que 5 reprovado

# entrada

notas = 0
bimestre = 1
soma = 0

while bimestre <= 4:
  notas = float (input(f"digite sia nota do {bimestre}° bimestre:  "))

  bimestre +=1
  soma = soma + notas

  if (soma/4)>= 7:
    print (f"a sua nota media final foi {soma/4}, você está aprovado. ")

  elif (soma/4)>= 5:
    print (f"a sua nota media final foi {soma/4}, você está de recuperação. ")

  elif (soma/4)<= 5:

    print (f"a sua nota media final foi {soma/4}, você está reprovado. ")
```
