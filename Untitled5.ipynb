```python
# escrerver um prograna que o usuário digita uma palavra e o programa verifica se a palavra é palíndroma e imprimi

palavra = str (input ("digite uma palavra"))
invertida = ""

for i in (palavra):
  invertida = i  + invertida

  if  palavra == invertida:
    print (f"{palavra} é palíndroma")

else:
  print (f"{palavra} não é palíndroma")
```

```python
# soma acumulada
#peça ao usuario para digitar um numero e imprimi o reultado da soma de 1 ate o primeiro numero digitado

try:
    numero = int(input("Digite um número inteiro positivo: "))

    if numero < 0:
        print("Digite um número positivo. ")

    else:
        soma_acumulada = 0

        for i in range(1, numero + 1):
            soma_acumulada += i
        print(f"A soma acumulada de 1 até {numero} é: {soma_acumulada}")

except ValueError:
    print("Entrada inválida. Por favor, digite um número inteiro.")
```
