```python
# 1 - contagem simples, o usuário digita um numero e  o programa imprimi do zero ate o numero

# entrada

n = int(input("Digite um numero: "))
for i in range (n):
  print(i)
```

```python
# 2 - contagem regressiva,o usuario digita um numero e o programa imprimi do numero ate o zero

# entrada

n = int(input("Digite um numero: "))
for i in range (n, -1, -1):
  print(i)
```

```python
# 3 - numeros pares, o usuario digita um numero e o programa imprimi os numeros pares

# entrada

n = int(input("Digite um numero: "))
for i in range (n):

  if i % 2 == 0:
    print(i)
```

```python
# 4 - numeros impares, o usuario digita um numero e o programa imprimi os numeros impares

# entrada

n = int(input("Digite um numero: "))
for i in range (n):
  if i % 2 != 0:
    print(i)
```

```python
# 5 - soma, o usuario digita um numero e o programa soma os numeros do zero até o numero

# entrada
n = int(input("Digite um numero: "))

# processamento
total = 0
for i in range(n + 1):
  total += i

# saida
print(f"A soma dos números de 0 até {n} é {total}")
```

```python
# 6 - tabuada, o usuario digita um numero e o programa imprimi a tabuada do numero

# entrada

n = int(input("Digite um numero: "))

print(f"Tabuada de {n}:")

for i in range(1, 11):

  print(f"{n} x {i} = {n * i}")
```

```python
# 7 - contar letras, o usuario digita uma palavra e o programa conta as letras da palavra

# entrada

palavra = input("Digite uma palavra/frase: ")


numero = len(palavra)

print(f"A palavra '{palavra}' tem {numero} letras")
```

```python
# 8 contar vogais, o usuario digita uma palavra e o  programa conta as vogais

# entrada

palavra = input("Digite uma palavra: ")


vogais = "aeiouAEIOU"
contador = 0
for letra in palavra:
  if letra in vogais:
    contador += 1

print(f"A palavra '{palavra}' tem {contador} vogais.")
```

```python
# 9 - maior numero digitado, o usuario digita quatro numeros e o programa imprimi o maior numero digitado

# entrada
n1 = float(input("Digite o primeiro número: "))
n2 = float(input("Digite o segundo número: "))
n3 = float(input("Digite o terceiro número: "))
n4 = float(input("Digite o quarto número: "))


maior = max(n1, n2, n3, n4)


print(f"O maior número digitado é: {maior}")
```

```python
# 10 - o usuario digita um numero entre 01 e 100 e o programa mostra quais sao divisíveis por 3 e 5

# entrada

n = int(input("Digite um numero entre 1 e 100: "))

for i in range(1, n + 1):
  if i % 3 == 0 and i % 5 == 0:
    print(i)
```
